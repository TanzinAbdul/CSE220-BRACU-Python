```python
array = [7, 11, 8, 2, 3, 15, 19, 21, 1]

class Node:
    
    def __init__(self, element, n = None):
        
        self.element = element
        self.next = n


class singly_cir_list:
    
    def __init__(self, array, start):
        
        
        self.start = start
        self.array = array
        self.cir = [0]*len(array)
        i = 0
        st = start
        for elem in array:
            self.cir[i] = array[st]
            i += 1
            st = (st+1)%len(self.cir)
            
        #swap code
        i = 1
        last = len(self.array)-1
        self.cir[i], self.cir[last] = self.cir[last], self.cir[i]
        
        #coverting to singly linked list
        self.head = None
        
        

        for elem in self.cir:
            if self.head == None:
                self.head = Node(elem)
                head = self.head
            else:
                node = Node(elem)
                head.next = node
                head = node
                
        head.next = self.head
        
    def max_head(self):
        max_idx = 0
        max_elem = self.cir[0]
        for elem in self.cir:
            if elem > max_elem:
                max_elem = elem
                
        for elem in self.cir:
            if elem == max_elem:
                break
            else:
                max_idx += 1
                
        i = max_idx
                
        self.head = Node(self.cir[i])
        head = self.head
        i = (i+1)%len(self.cir) 
        

        while  self.cir[i] != self.head.element:
            node = Node(self.cir[i])
            head.next = node
            head = node
            i = (i+1)%len(self.cir)
                
        head.next = self.head
        
    def display_array(self):
        print(self.cir)
        
    def display_list(self):
        
        head = self.head
        print(head.element, end= ', ')
        head = head.next
        while head.element != self.head.element:
            print(head.element, end =', ')
            head = head.next
        print()
        
        
temp = singly_cir_list(array, 4)
temp.display_list()
temp.max_head()
temp.display_list()
```
